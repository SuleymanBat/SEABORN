```python
#CORRELATIONS 
```

For data scientists, checking correlations is an important part of the exploratory data analysis process. This analysis is one of the methods used to decide which features affect the target variable the most, and in turn, get used in predicting this target variable. In other words, itâ€™s a commonly-used method for feature selection in machine learning.
And because visualization is generally easier to understand than reading tabular data, heatmaps are typically used to visualize correlation matrices. A simple way to plot a heatmap in Python is by importing and implementing the Seaborn library.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

#JOIN PLOT

A joinplot allows you to study the relationship between two numerical variables.The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot


```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#Pair plot
Also known as scatterplot in which one variable in the same data row is matched with another variable`s value, like this:Pairplots are just elaborates on this showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
df['sex'].value_counts()
```

# ** distplot******

It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
sns.countplot('sex',data=df)
```
